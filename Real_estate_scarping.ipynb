```python
import requests

url = "https://www.magicbricks.com/mbsrp/propertySearch.html"
querystring_base = {
    "editSearch": "Y",
    "category": "S",
    "propertyType": "10002,10003,10021,10022",
    "bedrooms": "11700,11701,11702,11703,11704",
    "city": "6403",
    "sortBy": "premiumRecent",
    "postedSince": "-1",
    "pType": "10002,10003,10021,10022",
    "isNRI": "N",
    "multiLang": "en"
}

payload = ""
headers = {
    "cookie": "your_cookie_here",
    "authority": "www.magicbricks.com",
    "accept": "application/json, text/plain, */*",
    "accept-language": "en-US,en;q=0.9",
    "cache-control": "no-cache",
    "pragma": "no-cache",
    "referer": "https://www.magicbricks.com/property-for-sale/residential-real-estate?bedroom=1,2,3,4,5&proptype=Multistorey-Apartment,Builder-Floor-Apartment,Penthouse,Studio-Apartment&cityName=Noida",
    "sec-ch-ua": "\"Chromium\";v=\"122\", \"Not(A:Brand\";v=\"24\", \"Microsoft Edge\";v=\"122\"",
    "sec-ch-ua-mobile": "?0",
    "sec-ch-ua-platform": "\"Windows\"",
    "sec-fetch-dest": "empty",
    "sec-fetch-mode": "cors",
    "sec-fetch-site": "same-origin",
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36 Edg/122.0.0.0"
}

all_items = []
total_results = 11483
results_per_page = 30
total_pages = (total_results + results_per_page - 1) // results_per_page  # Calculate total pages

# Looping through pages
for page in range(1, total_pages + 1):
    querystring = {**querystring_base, "page": str(page)}
    response = requests.request("GET", url, data=payload, headers=headers, params=querystring)
    result_json = response.json()
    result_items = result_json['resultList']
    
    # Checking for duplication and appending to all_items
    for item in result_items:
        if item not in all_items:
            all_items.append(item)

print("Total items:", len(all_items))
```

```python
import pandas as pd

# Extracting required attributes from all_items
data = []

for item in all_items:
    data.append({
        "encId": item.get('encId'),
        "url": item.get('url'),
        "price": item.get('price'),
        "sqFtPrice": item.get('sqFtPrice'),
        "bedroomD": item.get('bedroomD'),
        "carpetArea": item.get('carpetArea'),
        "carpAreaUnit": item.get('carpAreaUnit'),
        "devName": item.get('devName'),
        "locSeoName": item.get('locSeoName'),
        "prjname": item.get('prjname'),
        "ltcoordGeo": item.get('ltcoordGeo'),
        "possStatusD": item.get('possStatusD'),
        "OwnershipTypeD": item.get('OwnershipTypeD'),
        "transactionTypeD": item.get('transactionTypeD'),
        "floorNo": item.get('floorNo'),
        "floorD": item.get('floorD'),
        "balconiesD": item.get('balconiesD'),
        "bathD": item.get('bathD'),
        "furnishedD": item.get('furnishedD'),
        "facingD": item.get('facingD'),
        "adroomD": item.get('adroomD'),
        "availableFrom": item.get('availableFrom'),
        "isPrimeLocProp": item.get('isPrimeLocProp'),
        "luxAmenitiesD": item.get('luxAmenitiesD'),
        "luxAmenities": item.get('luxAmenities'),
        "sqFtPrD": item.get('sqFtPrD'),
        "propTypeD": item.get('propTypeD'),
        "propertyTitle": item.get('propertyTitle'),
        "ctName": item.get('ctName')
    })

 
```

```python
# Creating DataFrame
df = pd.DataFrame(data)
```
